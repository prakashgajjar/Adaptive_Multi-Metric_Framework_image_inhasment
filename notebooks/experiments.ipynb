## Experiment 1: Create a Sample Test Image

Generate a synthetic low-quality image to demonstrate enhancement

```python
# Create a synthetic test image with low contrast
def create_test_image(width=256, height=256, quality='low'):
    """
    Create a synthetic test image.
    
    Parameters:
    - width, height: Image dimensions
    - quality: 'low', 'medium', 'high' - determines image contrast
    """
    # Create gradient with geometric patterns
    x = np.linspace(0, 2*np.pi, width)
    y = np.linspace(0, 2*np.pi, height)
    X, Y = np.meshgrid(x, y)
    
    # Base pattern: combination of waves
    pattern = np.sin(X) * np.cos(Y) + np.sin(2*X) * np.cos(Y/2)
    
    # Normalize and add intensity modulation
    img = (pattern + 2) * 63.75  # Scale to roughly 0-255
    
    if quality == 'low':
        # Add noise and reduce contrast
        noise = np.random.normal(0, 15, img.shape)
        img = img + noise
        img = img * 0.6 + 50  # Darken and reduce range
    elif quality == 'medium':
        noise = np.random.normal(0, 10, img.shape)
        img = img + noise
    
    img = np.uint8(np.clip(img, 0, 255))
    
    # Convert grayscale to BGR (3 channels) for consistency
    img_bgr = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    
    return img_bgr

# Create test images with different quality levels
test_image_low = create_test_image(quality='low')
test_image_medium = create_test_image(quality='medium')
test_image_high = create_test_image(quality='high')

print("✓ Test images created")
print(f"  Image shape: {test_image_low.shape}")
print(f"  Data type: {test_image_low.dtype}")
print(f"  Low quality - Mean: {test_image_low.mean():.1f}, Std: {test_image_low.std():.1f}")
print(f"  High quality - Mean: {test_image_high.mean():.1f}, Std: {test_image_high.std():.1f}")
```

## Experiment 2: Apply Enhancement Techniques

Apply all four enhancement methods to the test image

```python
# Apply all enhancement techniques
original = test_image_low.copy()

enhancements = {
    'Histogram Eq.': apply_histogram_equalization(original),
    'CLAHE': apply_clahe(original),
    'Gamma Corr.': apply_gamma_correction(original),
    'Bilateral': apply_bilateral_filter(original)
}

print("✓ All enhancement methods applied")
print("\nEnhanced images created:")
for method, img in enhancements.items():
    print(f"  • {method}: shape={img.shape}, dtype={img.dtype}")
```
